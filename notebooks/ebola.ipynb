```python
from sympy import *
init_printing()
```

```python
q10, q01 =symbols("q_10 q_01")
betaCC, betaCH, betaHC, betaHH = symbols("beta_CC beta_CH beta_HC beta_HH")
gamma = symbols("gamma")
```

```python
eq1 = Eq(q10, (betaCC/(betaCC+betaCH+gamma))*q10**2 + (betaCH/(betaCC+betaCH+gamma))*q10*q01 + (gamma/(betaCC+betaCH+gamma)))
eq2 = Eq(q01, (betaHC/(betaHC+betaHH+gamma))*q10**2 + (betaHH/(betaHC+betaHH+gamma))*q10*q01 + (gamma/(betaHC+betaHH+gamma)))
display(eq1)
display(eq2)
```

```python
solve([eq1,eq2], [q10,q01])
```
