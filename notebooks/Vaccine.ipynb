```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 16
```

```python
N = 100_000
mu = 1/8 # Recovery rate
Nv = 50_000 # 70% of the population were vaccinated
alpha = 0.2
```

```python
def coverage(mu,alpha,Nt,Nv):
    Pa=[]
    for i in np.arange(0.2,6,0.1):
        R_0=i
        beta = R_0*mu/(N+Nv)
        q = mu/(Nt*beta-Nv*alpha*beta+beta*Nv)
        p = 1-q
        if p <0: 
            Pa.append(0)
        else:
            Pa.append(p)

    return Pa
```

```python
Pa_11 = coverage(mu,0.2,0.7*N,0.3*N)
Pa_21 = coverage(mu,0.2,0.5*N,0.5*N)
Pa_31 = coverage(mu,0.2,0.3*N,0.7*N)
```

```python
R_0=np.arange(0.2,6,0.1)
plt.plot(R_0,Pa_11,marker='o', label = '$30\%$ coverage')
plt.plot(R_0,Pa_21,marker='o', label = '$50\%$ coverage')
plt.plot(R_0,Pa_31,marker='o', label = '$70\%$ coverage')
plt.xlabel('$R_{0}$ values')
plt.ylabel('Probability')
plt.title('Probability of major outbreak $\\alpha=0.2$')
plt.legend()
plt.show()
```
