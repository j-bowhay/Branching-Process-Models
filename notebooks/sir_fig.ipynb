```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science','ieee'])

N = 500
beta = 0.3
mu = 0.1

def rhs(t, y):
    S = y[0]
    I = y[1]
    R = y[2]
    return [-beta*I*S/N, beta*I*S/N - mu*I, mu*I]

sol = solve_ivp(rhs, t_span=(0, 100), y0=[N-10, 10, 0], rtol=1e-9)

fig, ax = plt.subplots()
ax.plot(sol.t, sol.y[0, :], label="S")
ax.plot(sol.t, sol.y[1, :], label="I")
ax.plot(sol.t, sol.y[2, :], label="R")
ax.set_xlim([0, 100])
ax.set_xlabel("$t$")
ax.set_ylabel("\# People")
ax.legend()
plt.savefig("figures/sir.pdf")
```
